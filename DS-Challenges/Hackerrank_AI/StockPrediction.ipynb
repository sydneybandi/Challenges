```python
import numpy as np
import pandas as pd
import wget
import matplotlib.pyplot as plt
```

```python
train_dir = wget.download("http://s3.amazonaws.com/hr-testcases-us-east-1/329/assets/bigger_data_set.txt")
datafile = open(train_dir, "r")
data = []

for line in datafile:
    data.append([i for i in line.strip().split(' ')])

df = pd.DataFrame(data)
df = df.set_index(0).T
df = df.astype(float)
df.head()
```

```python
df.plot(figsize=(15,5), lw = 2, colormap='jet', title = "Stock values over time")
plt.xlabel('Days')
plt.ylabel('Stock Price')
plt.show()
```
